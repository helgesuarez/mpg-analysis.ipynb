```python
import seaborn as sns
```

```python
sns.load_dataset("mpg")
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.scatterplot(data=mpg, x="horsepower", y="mpg")

plt.title("Horsepower vs Fuel Economy")
plt.xlabel("Horsepower")
plt.ylabel("Miles Per Gallon (MPG)")

plt.show()
```

```python
# when horsepower goes up, the MPG goes down
# this would mean that there is an inverse relationship/correlation between the two variables
# the higher the horsepower of vehicles, the less MPG vehicles uses
```

```python
mpg_by_origin = mpg.groupby("origin")["mpg"].mean().reset_index()
mpg_by_origin
```

```python
# Japan has the highest use of mpg out of all of them
# Europe has the second highest one, whilst the USA has the lowest.
```

```python
plt.figure()
sns.lineplot(data=mpg, x="model_year", y="mpg")

plt.title("Fuel Economy Over Time")
plt.xlabel("Model Year")
plt.ylabel("MPG")

plt.show()
```

```python
# the MPG started increasing again in 1973 and onwards after having a dip from 1971-1972, most likely due to politics.
```

```python
plt.figure()
sns.boxplot(data=mpg, x="origin", y="mpg")

plt.title("Fuel Economy by Origin")
plt.xlabel("Origin")
plt.ylabel("MPG")

plt.show()
```

```python
# Japan has the highest median income in Europe and usa
# But the USA has the longest range of all, compared to Japan and Europe
```

```python
plt.figure()
sns.scatterplot(data=mpg, x="weight", y="mpg")

plt.title("Weight vs Fuel Economy")
plt.xlabel("Weight")
plt.ylabel("MPG")

plt.show()
```

```python
# The bigger the weight of a vehicle, the lower the MPG
# This is another example of an inverse relationship between the two variables
```
